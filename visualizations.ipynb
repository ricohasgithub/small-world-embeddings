Hyperblic Embeddings

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

def embedding(point):
    return -np.sqrt(0.3*(point[0]**2 + point[1]**2) + 1)

X = np.linspace(-5, 5, 100)
Y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(X, Y)

Z = np.sqrt(0.3*(X **2 + Y **2) + 1)

V = [[0, 0], [2, 2], [-2, 2], [2, -2], [-2, -2]]
embedded = [[point[0], point[1], embedding(point)] for point in V]
print(embedded)

# Create a figure for plotting
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# ax.plot_surface(X, Y, Z, cmap='jet', alpha=0.6, vmin=-5, vmax=5)
ax.plot_surface(X, Y, -Z, cmap='jet', alpha=0.6, vmin=-5, vmax=5)

for point in embedded:
    ax.scatter(point[0], point[1], point[2], color='red')
```

Hyperspherical Embeddings

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# draw sphere
u, v = np.mgrid[0:2*np.pi:25j, 0:np.pi:25j]
x = np.cos(u)*np.sin(v)
y = np.sin(u)*np.sin(v)
z = np.cos(v)

# Create a figure for plotting
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# alpha controls opacity
ax.plot_surface(x, y, z, cmap='jet', alpha=0.6)
ax.scatter(0, -1, 0, color='black')
ax.scatter(0, 1, 0, color='black')
ax.scatter(1, 0, 0, color='black')
ax.scatter(-1, 0, 0, color='black')
```
